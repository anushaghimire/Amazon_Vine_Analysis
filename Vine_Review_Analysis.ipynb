```python
import pandas as pd
```

```python
#Read in the vine_table.csv file as a DataFrame
vine_df = pd.read_csv("vine_table.csv")
vine_df.dtypes
```

```python
vine_df.count()
```

```python
#Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20
vine_table = vine_df[vine_df.total_votes >= 20]
vine_table
```

```python
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
vine_table_df = vine_table[(vine_table.helpful_votes/vine_table.total_votes) >= 0.50]
vine_table_df
```

```python
#Filter the DataFrame and create a new DataFrame that retrieves all the rows where a review was written as part of 
#the Vine program (paid)
vine_reviews_paid = vine_table_df[vine_table_df.vine == 'Y']
vine_reviews_paid
```

```python
#retrieve all the rows where the review was not part of the Vine program (unpaid)
vine_reviews_unpaid = vine_table_df[vine_table_df.vine == 'N']
vine_reviews_unpaid
```

```python
#Determine the total number of reviews(paid)
total_paid_reviews = vine_reviews_paid.vine.count()
total_paid_reviews
```

```python
# Total number of 5-star reviews
paid_5_star = vine_reviews_paid[vine_reviews_paid.star_rating == 5].vine.count()
paid_5_star
```

```python
#percentage of 5-star reviews (paid)
percentage_5_star_reviews =(paid_5_star/total_paid_reviews)*100
percentage_5_star_reviews
```

```python
total_unpaid_reviews = vine_reviews_unpaid.vine.count()
total_unpaid_reviews
```

```python
unpaid_5_star = vine_reviews_unpaid[vine_reviews_unpaid.star_rating == 5].vine.count()
unpaid_5_star
```

```python
#percentage of 5-star reviews (unpaid)
percentage_5_star_reviews = (unpaid_5_star/total_unpaid_reviews) *100 
percentage_5_star_reviews
```
